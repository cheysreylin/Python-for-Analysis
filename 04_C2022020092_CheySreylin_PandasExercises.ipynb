### 2. Get all the items of series A that doesn't exist in series B

```python
s_A = pd.Series([1,3,5,7,9])
s_B = pd.Series([1,2,3,4,5])

result = s_A[-(s_A.isin(s_B))]
print(result)
```

### 12. Join dataframes df1 and df2 by ‘fruit-pazham’ and ‘weight-kilo’. Make sure that they have only the common rows.

```python
df1 = pd.DataFrame({'fruit': ['apple', 'banana', 'orange'] * 3,
                    'weight': ['high', 'medium', 'low'] * 3,
                    'price': np.random.randint(0, 15, 9)})

df2 = pd.DataFrame({'pazham': ['apple', 'orange', 'pine'] * 2,
                    'kilo': ['high', 'low'] * 3,
                    'price': np.random.randint(0, 15, 6)})

res = df1.merge(df2,how = 'inner',indicator = True)
print(res)
```
